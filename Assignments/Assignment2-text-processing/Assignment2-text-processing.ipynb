### Question 1. (2 points)

Write a function that counts the frequency of all the [palindrome](https://en.wikipedia.org/wiki/Palindrome) words in a text file and test it on `data/melville-md.txt`. 

- Requirement: word counting should be **case-insensitive**

````python
# your code here
def palindromes_frequencies(file):
    """
    This function counts the frequencies of palindrome words in a text file.
    """
    infile = open(file, "r", encoding = "utf8")
    alltext = infile.read().lower()
    
    import re
    
    palindromes_counts = dict()
    for word in re.split(r"\W+",alltext):
        if re.search(r"\D",word) and len(word) > 2 and word == word[::-1]:
            if word in palindromes_counts:
                palindromes_counts[word] += 1
            else:
                palindromes_counts[word] = 1
        
    print("\n***sorting frequencies***\n")
    

    for word, fr in sorted(palindromes_counts.items(), key = frequencies, reverse = True):
        print(word, fr, end = "|", sep = ":")
        
    infile.close()
    
melville =  "data/melville-md.txt"
palindromes_frequencies(melville)
````

````python
def original(input_file):
    stripped_words = set()
    infile = open(input_file, "r", encoding = "utf8")
    for line in infile:
        splLine= line.split()
        for word in splLine:
            consonants = "".join([c for c in word.lower() if c not in ["a","e","i","o","u"]])
            stripped_words.add(consonants)
    infile.close()
````

````python
# your code here
def consonants_sequences_extractor_1(input_file):
    """
    This function extracts consonants sequences 
    from a text file to add them to a sorted list.
    """

    infile = open(input_file, "r", encoding = "utf8")
    alltext = infile.read().lower()
    
    consonants_set = set()
    
    for word in re.split(r"\W+",alltext):
        consonants = "".join([c for c in word if c not in ["a","e","i","o","u"] and not c.isdigit()])
        consonants_set.add(consonants) 
        
    consonants_sequences = (list(consonants_set))
    consonants_sequences.sort()
    print(consonants_sequences)
    
    infile.close()       

    
consonants_sequences_extractor_1(melville)
                                                                
````

````python
# your code here
def consonants_sequences_extractor_2(input_file):
    """
    This function extracts consonants sequences  
    from a text file to add them to a sorted list.
    """

    infile = open(input_file, "r", encoding = "utf8")
    alltext = infile.read().lower()
    
    consonants_set = set()
    for word in re.split(r"\W+",alltext):
        consonants = "".join([c for c in word if re.search(r"[^aeiou^0-9]",c)])
        consonants_set.add(consonants)
        
    consonants_sequences = (list(consonants_set))
    consonants_sequences.sort()
    print(consonants_sequences)
    
    infile.close()
    
            
consonants_sequences_extractor_2(melville)
                                                                
````

````python
# your code here
def consonants_sequences_extractor_3(input_file):
    """
    This function extracts consonants sequences  
    from a text file to add them to a sorted list.
    """

    infile = open(input_file, "r", encoding = "utf8")
    alltext = infile.read().lower()
    
    consonants = ["".join([c for c in word if re.search(r"[^aeiou^0-9]",c)]) for word in re.split("\W+",alltext)]
    consonants_set = set()
    for word in consonants:
        consonants_set.add(word)
        
    consonants_sequences = (list(consonants_set))
    consonants_sequences.sort()
    print(consonants_sequences)
                
    
consonants_sequences_extractor_3(melville)
````
